非线性激活函数让神经元变成“可调形状的小基函数”；很多这样的基函数加起来，就能拼出复杂函数。

真正起作用的不是“ReLU 这个具体公式”，而是：
这个激活函数 𝜎 能不能生成一个“足够丰富的函数族”。

只要𝜎不是太退化（比如纯线性），通常就能通过𝜎(𝑤𝑥+𝑏) 这些变形，产生很多不同形状，再通过线性组合逼近复杂函数。

即使不是 ReLU，只要激活函数能构造出“局部响应”或者“分段/波包式”的成分，也可以用类似思想逼近：
sigmoid：像平滑台阶
tanh：像平滑 S 形
Gaussian（如果拿来当激活/基函数）：像局部小包
ReLU：像折线基元

它们都能通过叠加形成复杂形状，只是“拼法”和效率不同。

你可以把神经网络看成在构造这个集合：
$F_m=\left\{\sum_{j=1}^{m} a_j \sigma(w_j x+b_j)+c\right\}$

当 𝑚（神经元数）越来越大时，集合变得越来越大。

我们关心的是：这些集合的并集 $\bigcup_{m=1}^{\infty} F_m$在连续函数空间里是否“稠密”（dense）。



ReLU（激活）逐元素（N*D）个元素全部保留>0的，不跨样本，不跨特征做统计

BatchNorm（训练时）对每个特征维度 𝐷 在 batch 维𝑁上算均值方差

也就是“按列统计，再逐元素归一化”

Softmax（分类输出时）通常对每个样本的类别维度做（按行）不是对整个矩阵所有元素一起做

在单隐层、紧集上一致逼近连续函数的设定下，并非任意非线性激活函数都具有万能逼近性质。对连续激活函数，Leshno 等（1993）给出的必要充分条件是：激活函数必须是非多项式。若激活函数是多项式，则该函数类不具备稠密性，因此不能万能逼近。https://archive.nyu.edu/bitstream/2451/14384/1/IS-91-26.pdf
